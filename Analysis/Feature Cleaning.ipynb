### Concat Dataframes

```python
data = pd.read_csv('Amazon_earbuds.csv')
data2 = pd.read_csv('Amazon_earbuds2.csv')
data3 = pd.read_csv('Amazon_earbuds3.csv')
data = pd.concat([data,data2,data3],ignore_index=True)
data
```

```python
data = data.drop(data.columns[0],axis=1)
data
```

### Brands

```python
data['Brands'] = data['Brands'].str.replace('Visit the ','')
data['Brands'] = data['Brands'].str.replace('Store','')
data['Brands'] = data['Brands'].str.replace('Brand: ','')
data['Brands'] = data['Brands'].str.strip()
data['Brands'].unique()
```
