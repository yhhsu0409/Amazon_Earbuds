```python
import pandas as pd
```

```python
data = pd.read_csv('Clean_Amazon_Earbuds.csv')
data
```

```python

import ast 

for index,strings in data['Review_Words'].items():
    data['Review_Words'][index] = ast.literal_eval(strings) 
data['Review_Words'][0]
```

```python
Words = []
for index,words in data['Review_Words'].items():
    for word in words:
        if word == 'No Review':
            continue
        else:
            Words.append(word)
Words
```

```python
from collections import Counter
count = Counter(Words)
count
```

```python
x = {}
for i,j in count.items():
    x[i]=j
x
```

```python
index = [y for y in range(len(x))]

words_data = pd.DataFrame.from_dict(x,orient='index')

words_data
```

```python
words = words_data.index
```

```python
words_data.reset_index(inplace=True)
words_data.rename(columns={"index": "Words", 0: "Counts"},inplace=True)
```

```python
words_data.sort_values('Counts',ascending=False)
```

```python
words_data.to_csv('Words_Data.csv')
```
