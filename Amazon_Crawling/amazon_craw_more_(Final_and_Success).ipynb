```python
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.common.exceptions import NoSuchElementException
import sys
import time 
```

```python
path = '/Users/jack/Desktop/Python/chromedriver'
driver = webdriver.Chrome(executable_path=path)
```

```python
driver.get('https://www.amazon.com')
driver.maximize_window()
driver.implicitly_wait(3)
```

```python
link_urls = []
for i in range(1,21):
    while i <20:
        amazon = BeautifulSoup(driver.page_source)
        driver.implicitly_wait(5)
        links = amazon.find_all('a','a-link-normal a-text-normal')
        for link in links:
            url =link['href']
            link_urls.append(url)
        next_list = driver.find_elements_by_partial_link_text('Next')
        next_list[-1].click()
        time.sleep(3)
    else:
        amazon = BeautifulSoup(driver.page_source)
        time.sleep(3)
        links = amazon.find_all('a','a-link-normal a-text-normal')
        for link in links:
            url =link['href']
            link_urls.append(url)
```

```python
#check_duplicates
def checkIfDuplicates_2(listOfElems):
    ''' Check if given list contains any duplicates '''    
    setOfElems = set()
    for elem in listOfElems:
        if elem in setOfElems:
            return True
        else:
            setOfElems.add(elem)         
    return False
checkIfDuplicates_2(link_urls)
```

```python
titles = []
brands = []
prices = []
stars = []
urls = []
number_ratings = []
review_words =[]
no_reviews_items = []
```

```python
for url in range(len(link_urls)):
    js = 'window.open("%s");'%link_urls[url]
    driver.execute_script(js)
    driver.switch_to.window(driver.window_handles[1])
    driver.implicitly_wait(3)
    ### Title
    try:
        title = driver.find_element_by_xpath('//*[@id="productTitle"]').text
    except NoSuchElementException:
        driver.close()
        driver.switch_to.window(driver.window_handles[0])
        continue
    titles.append(title)
    ### Brand
    brand = driver.find_element_by_xpath('//*[@id="bylineInfo"]').text
    brands.append(brand)
    ### Ratings
    try:
        number_rating = driver.find_element_by_xpath('//*[@id="acrCustomerReviewText"]').text
        number_ratings.append(number_rating)
    except NoSuchElementException:
        number_ratings.append('No Ratings')
    ### Price
    try:
        price = driver.find_element_by_xpath('//*[@id="price_inside_buybox"]').text
    except NoSuchElementException:
        try:
            price = driver.find_element_by_xpath('//*[@id="newBuyBoxPrice"]').text
        except NoSuchElementException:
            try:
                price = driver.find_element_by_xpath('//*[@id="priceblock_dealprice"]').text
            except NoSuchElementException:
                try:
                    price = 'No Price'
                except NoSuchElementException:
                    sys.exit("Error message")
    prices.append(price)
    
    ### Reviews
    for i in range(10):
        js = "window.scrollTo(0,%s)"%(3000*i)
        driver.execute_script(js)
        driver.implicitly_wait(7)
        try:
            reviews = driver.find_elements_by_class_name('cr-lighthouse-term')
        except NoSuchElementException:
            continue
        if len(reviews)>0:
            break
 ### Stars
    try:
        star = driver.find_element_by_xpath('//*[@id="reviewsMed ley"]/div/div[1]/div[2]/div[1]/div/div[2]/div/span/span').text
    except NoSuchElementException:
        try:
            star = driver.find_element_by_xpath('//*[@id="reviewsMedley"]/div/div[1]/div[2]/div[1]/div/div[2]/div/span/span').text
        except NoSuchElementException:
            try:
                star = 'No Review'
            except NoSuchElementException:
                sys.exit("Error message")
    stars.append(star)
       
    if len(reviews)==0:
        reviews = 'No Review'
    review_word = []
    for i in range(len(reviews)):
        if reviews == 'No Review':
            review_word.append('No Review')
            no_reviews_items.append(review_word)
            break
        if reviews[i].text == '':
            break
        review_word.append(reviews[i].text)
    review_words.append(review_word)
    
    ### Url
    urls.append(link_urls[url])
    driver.close()
    driver.switch_to.window(driver.window_handles[0])
    time.sleep(2)
```

```python
print('Count of Titles %d'%len(titles))
print('Count of Stars %d'%len(stars))
print('Count of Price %d'%len(prices))
print('Count of Brands %d'%len(brands))
print('Count of Ratings %d'%len(number_ratings))
print('Count of Reviews %d'%len(review_words))
print(titles)
print(stars)
print(prices)
print(brands)
print(number_ratings)
print(review_words)
```
